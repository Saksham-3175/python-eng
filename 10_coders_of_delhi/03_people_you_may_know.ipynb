```python
import json

def load_data(filename):
    with open(filename, 'r') as f:
        return json.load(f)

def mutuals(user_id, data):
    user_friends = {}
    
    for user in data['users']:
        user_friends[user['id']]=set(user['friends'])
        print(f"user_friends={user_friends}")
    
    #Sanity Check
    if user_id not in user_friends:
        return []
    
    direct_friends = user_friends[user_id]
    print(f"direct_friends={direct_friends}")
    suggestions = {}

    for friend in direct_friends:
        for mutual in user_friends[friend]:
            if mutual!=user_id and mutual not in direct_friends:
                #Counting the mutuals
                suggestions[mutual] = suggestions.get(mutual, 0) + 1
    print(f"suggestions= {suggestions}")
    sorted_suggestions = sorted(suggestions.items(), key=lambda x:x[1], reverse=True)
    print(f"Sorted Suggestions = {sorted_suggestions}")
    return [user_id for user_id, mutual_count in sorted_suggestions]
    

```

```python
#Loading the data
data = load_data('massive_data.json')
user_id = 2
recc = mutuals(user_id, data)
print(recc)
```
